```python
# Import libraries
import pandas as pd
import numpy as np
import py_stringmatching
```

```python
# Read ClinicalTrials.gov data
df_clin = pd.read_csv('ClinicalTrialsGov-Pacemaker.csv')
df_clin.head()
```

```python
# Read NIH data
df_nih = pd.read_excel('pacemaker_data.xlsx')
df_nih.head()
```

```python
def lookformatch(nih,df_clin):
    '''Try to match up nih project with a clinical trial from df_clin'''
    # Filter by organization/university
    df_result = df_clin[[str(nih['ORG_NAME']).title() in str(o).title() for o in df_clin['Sponsor/Collaborators'].to_list()]]
    
    # Filter by project title - should replace with better/more complex solution
    jac = py_stringmatching.Jaccard()
    ws_tok = py_stringmatching.WhitespaceTokenizer()
    df_result = df_result.loc[[jac.get_sim_score(ws_tok.tokenize(nih['PROJECT_TITLE']),t)>0.2 for t in df_result['Title_tok'].to_list()]]
    
    return df_result

ws_tok = py_stringmatching.WhitespaceTokenizer()
df_clin['Title_tok'] = df_clin.apply(lambda row: ws_tok.tokenize(row['Title']),axis=1)

# Try to find matches first by organization and then by similarity in titles
matches = []
for i in range(len(df_nih)):
    test = df_nih.loc[i]

    df_result = lookformatch(test,df_clin)  
    if len(df_result)>0:
        print(i)
        matches.append([i,df_result.index.to_list()])
        
print(df_result['Title'].to_list())
print(test['PROJECT_TITLE'])
```

```python
# Print titles of matches to confirm
for i in range(len(matches)):
    idx_nih = matches[i][0]
    idx_clin = matches[i][1]
    print(df_clin.loc[idx_clin]['Title'].to_list())
    print(df_nih.loc[idx_nih]['PROJECT_TITLE'])
    print('')
```
